```python
def cicle_games(df, ano_copa):

    """
    Filtra todas os jogos por ciclo, porém apenas jogos que alguma seleção classificada para
    a copa jogou.
    """

    if ano_copa != 2022 and ano_copa != 2026:
        inicio_ciclo = date(year=ano_copa-4, month=8, day=1)
        fim_ciclo = date(year=ano_copa, month=5, day=26)
    
    elif ano_copa == 2022:
        inicio_ciclo = date(year=ano_copa-4, month=8, day=1)
        fim_ciclo = date(year=ano_copa, month=11, day=19)

    elif ano_copa == 2026:
        inicio_ciclo = date(year=ano_copa-4, month=12, day=20)
        fim_ciclo = date(year=ano_copa, month=6, day=10)

    selecoes = pd.read_csv(f'dados/copas/FIFA - {ano_copa}.csv')
    selecoes = selecoes['Team']
    jogos = df
    df_ciclo = jogos[(jogos['date'] >= inicio_ciclo) & (jogos['date'] <= fim_ciclo)]
    df_ciclo = df_ciclo[(df_ciclo['home_team'].isin(selecoes)) | (df_ciclo['away_team'].isin(selecoes))]
    df_ciclo["resultado"] = np.where(
        df_ciclo["home_score"] > df_ciclo["away_score"], 1, 
        np.where(df_ciclo["away_score"] > df_ciclo["home_score"], 2, 0)  
    )
    #!!!!!
    # 1 = Time da casa venceu ; 2 = Time de fora venceu ; 0 = Empate
    df_ciclo = df_ciclo.reset_index()
    df_ciclo.to_csv(f"dados/confrontos/historico_confrontos_ciclo{ano_copa}.csv")

for ano in anos_copa:
    try:
        cicle_games(df, ano)
    except: ...
```

```python
def stats_cicle(ano):
    """
    Gera tabelas de classificação baseadas nos jogos do ciclo, contando vitórias, derrotas, empates,
    e pontuando as seleções.
    """
    if ano == 2026:
        teams = [normalize_team_name(t) for t in qualified_teams]
    else:
        teams = list(set(pd.read_csv(f"dados/copas/FIFA - {ano}.csv")["Team"]))
        teams = [normalize_team_name(t) for t in teams]

    df_confrontos = pd.read_csv(f"dados/confrontos/historico_confrontos_ciclo{ano}.csv")
    all_teams = pd.concat([df_confrontos['home_team'], df_confrontos['away_team']]).unique()
    mapa_nomes = {team: normalize_team_name(team) for team in all_teams}
    
    df_confrontos['home_team'] = df_confrontos['home_team'].map(mapa_nomes)
    df_confrontos['away_team'] = df_confrontos['away_team'].map(mapa_nomes)

    print(len(teams))
    print(teams)

    df_stats = pd.DataFrame({"team": teams})
    df_stats = df_stats.set_index("team")
    df_stats["wins"] = 0
    df_stats["draws"] = 0
    df_stats["loses"] = 0
    df_stats["points"] = 0
    df_stats["games_played"] = 0
    df_stats["goals_for"] = 0
    df_stats["goals_against"] = 0

    for row in df_confrontos.itertuples():
        home = row.home_team
        away = row.away_team
        home_score, away_score = row.home_score, row.away_score

        # só soma se o time estiver na lista da Copa
        is_home_qualified = home in df_stats.index
        is_away_qualified = away in df_stats.index

        if is_home_qualified: 
            df_stats.at[home, "games_played"] += 1
            df_stats.at[home, "goals_for"] += home_score     
            df_stats.at[home, "goals_against"] += away_score

        if is_away_qualified: 
            df_stats.at[away, "games_played"] += 1
            df_stats.at[away, "goals_for"] += away_score    
            df_stats.at[away, "goals_against"] += home_score
        
        if home_score > away_score:
            if is_home_qualified: df_stats.at[home, "wins"] += 1
            if is_away_qualified: df_stats.at[away, "loses"] += 1

        elif away_score > home_score:
            if is_home_qualified: df_stats.at[home, "loses"] += 1
            if is_away_qualified: df_stats.at[away, "wins"] += 1
            
        else:
            if is_home_qualified: df_stats.at[home, "draws"] += 1
            if is_away_qualified: df_stats.at[away, "draws"] += 1

    df_stats["goal_difference"] = df_stats["goals_for"] - df_stats["goals_against"]
    df_stats["goal_diff_per_game"] = round(df_stats["goal_difference"] / df_stats["games_played"], 2)
    df_stats["points"] = df_stats["wins"]*3 + df_stats["draws"]
    df_stats["points_per_game"] = round(df_stats["points"] / df_stats["games_played"],2)
    df_stats = df_stats.sort_values(by="points", ascending=False)
    df_stats.to_csv(f"dados/estatisticas_ciclos/estatisticas_ciclo{ano}.csv")

for ano in anos_copa:
    stats_cicle(ano)
```

```python
#Separando os csv que serão usados para treinar o modelo!


for ano in anos_copa:
    df_confrontos = pd.read_csv(f"dados/confrontos_copas/confrontos{ano}.csv")
    df_ciclo = pd.read_csv(f"dados/estatisticas_ciclos/estatisticas_ciclo{ano}.csv")
    
    df_ciclo["team"] = df_ciclo["team"].apply(normalize_team_name)
    df_ciclo = df_ciclo.set_index("team")
    
    df_input_modelo = pd.DataFrame()

    wins_difference = []
    draws_difference = []
    loses_difference = []
    points_per_game_difference = []
    goal_diff_pg_difference = []
    allign_results = []

    df_ciclo["win_rate"] = (df_ciclo["wins"] / df_ciclo["games_played"]).fillna(0)
    df_ciclo["draw_rate"] = (df_ciclo["draws"] / df_ciclo["games_played"]).fillna(0)
    df_ciclo["lose_rate"] = (df_ciclo["loses"] / df_ciclo["games_played"]).fillna(0)

    # Padroniza todas as colunas: minúsculas e com '_' no lugar do espaço
    df_confrontos.columns = df_confrontos.columns.str.lower().str.replace(' ', '_')

    unmatched_teams = set()
    for row in df_confrontos.itertuples():
        home = normalize_team_name(row.home_team_name)
        away = normalize_team_name(row.away_team_name)
        if home not in df_ciclo.index:
            unmatched_teams.add(home)
        if away not in df_ciclo.index:
            unmatched_teams.add(away)
    
    if unmatched_teams:
        print(f"Aviso Ano {ano}: Seleções não encontradas em estatísticas: {unmatched_teams}")

    for row in df_confrontos.itertuples():
        home_team = normalize_team_name(row.home_team_name)
        away_team = normalize_team_name(row.away_team_name)
        
        # Pular linhas com equipes não mapeadas
        if home_team not in df_ciclo.index or away_team not in df_ciclo.index:
            continue

        wins_difference.append(df_ciclo.loc[home_team]["win_rate"] - df_ciclo.loc[away_team]["win_rate"])
        draws_difference.append(df_ciclo.loc[home_team]["draw_rate"] - df_ciclo.loc[away_team]["draw_rate"])
        loses_difference.append(df_ciclo.loc[home_team]["lose_rate"] - df_ciclo.loc[away_team]["lose_rate"])
        points_per_game_difference.append(df_ciclo.loc[home_team]["points_per_game"] - df_ciclo.loc[away_team]["points_per_game"])
        goal_diff_pg_difference.append(df_ciclo.loc[home_team]["goal_diff_per_game"] - df_ciclo.loc[away_team]["goal_diff_per_game"])
        allign_results.append(row.resultado)
    
    df_input_modelo["wins_difference"] = wins_difference
    df_input_modelo["draws_difference"] = draws_difference
    df_input_modelo["loses_difference"] = loses_difference
    df_input_modelo["points_per_game_difference"] = points_per_game_difference
    df_input_modelo["result"] = allign_results
    df_input_modelo['goal_diff_pg_difference'] = goal_diff_pg_difference

    total_original = len(df_confrontos)
    total_processado = len(df_input_modelo)
    print(f"Ano {ano}: {total_processado} de {total_original} linhas processadas")
    df_input_modelo.to_csv(f"dados/dados_modelo/input_modelo{ano}.csv", index=False)
```
